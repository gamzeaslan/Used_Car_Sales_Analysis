```python
dataFrame=dataFrame.drop("transmission",axis=1)
```

```python
#veri temizlendi ve düzenlendi artık model oluşturma sürecine başlayabiliriz
```

```python
x=dataFrame.drop("price",axis=1).values
y=dataFrame["price"].values
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y
```

```python
#data=test+train
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.shape
```

```python
#fit the data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
#Creating a model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

#creaiting layers
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,batch_size=250,epochs=300)
```

```python
lossData=pd.DataFrame(model.history.history)
```

```python
lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_squared_error ,mean_absolute_error
```

```python
prediction_sequence=model.predict(x_test)
```

```python
prediction_sequence
```

```python
mean_absolute_error(y_test,prediction_sequence)
```
